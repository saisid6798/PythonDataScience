```python
#Read Total profit of all months and show it using a line plot
months=np.array(sales['month_number'])
total_profit=np.array(sales['total_profit'])
total_units=np.array(sales['total_units'])

plt.plot(months, total_profit,color='green',linewidth='4',alpha=1, marker='o', markersize=6 ,linestyle='dashed',
         label='Profit data of last year')
plt.xlabel("Month Number",size=10)
plt.ylabel("Total Profit",size=10)
plt.title("Comapny Profit data of last year",size=15)
plt.legend()
plt.show()
```

```python
#Read toothpaste sales data of each month and show it using a scatter plot
#Also, add a grid in the plot

toothpaste=np.array(sales['toothpaste'])
months=np.array(sales['month_number'])
plt.scatter(months, toothpaste, color='blue',alpha=0.4, s=100,label='Toothpaste sales')
plt.legend()
plt.grid()
plt.show()
```

```python
#Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

bathingsoap= np.array(sales['bathingsoap'])
months=np.array(sales['month_number'])
plt.bar(months,bathingsoap,align='center',edgecolor='green',width=0.5,color='red')
plt.xlabel("Month Number")
plt.ylabel("Sales of Bathing Soap")
plt.title("Sales of Bathing Soap Month Wise")
plt.legend()
plt.show()
plt.savefig("C:\\Users\\Sai Sidharth\\Downloads\\Assignment_Matplotlib1.png")
```

```python
#Calculate total sale data for last year for each product and show it using a Pie chart.your pie chart output should like this.
total_sales=sales[['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']].sum()

total_sales.to_numpy

total_sales_plot=total_sales
labels=["Face Cream","Face Wash","ToothPaste","Bathing Soap","Shampoo","Moisturizer"]
colors=['black','hotpink','red','green','blue','#4CAF50']
plt.pie(total_sales_plot,labels=labels)
plt.legend()
plt.show()
```

```python
#Read Bathing soap ,facewash of all months and display it using the Subplot.your output should be like this as mention below:
bathingsoap=np.array(sales['bathingsoap'])
months=np.array(sales['month_number'])

plt.subplot(2,1,1)
plt.plot(months, bathingsoap,color='black',linewidth='4',alpha=0.8,marker='o',markersize=6,linestyle='dashed',
        label='Sales Data of Bathing Soap')
plt.xlabel("Month Number",size=10)
plt.ylabel("Sales units in number",size=10)
plt.title("Company sales data", size=20)

facewash=np.array(sales['facewash'])
months=np.array(sales['month_number'])

plt.subplot(2,1,2)
plt.plot(months, facewash,color='red',linewidth='4',alpha=0.8,marker='o',markersize=6,linestyle='dashed',
        label='Sales Data of Bathing Soap')
plt.xlabel("Month Number",size=10)
plt.ylabel("Sales units in number",size=10)
plt.title("Company sales data", size=20)

plt.show()
```

```python
#Read all product sales data and show it using the stack plot
months=np.array(sales['month_number'])
facecream=np.array(sales['facecream'])
facewash=np.array(sales['facewash'])
toothpaste=np.array(sales['toothpaste'])
bathingsoap=np.array(sales['bathingsoap'])
shampoo=np.array(sales['shampoo'])
moisturizer=np.array(sales['moisturizer'])

plt.stackplot(months, facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer,
             colors=['red','blue','green','black','orange','brown'], labels=["Face Cream","Face Wash","Tooth Paste","Bathing Soap",
                                                                            "Shampoo","Moisturizer"])

plt.xlabel("Month Number",size=12)
plt.ylabel("Sales Units in Number",size=12)
plt.title("All Product Sales using Stack Plot")
plt.legend()
plt.show()
```
